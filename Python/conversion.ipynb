```python
def randtime():
    p= int(np.random.random()*24)
    q=int(np.random.random()*60)
    p=str(p)+":"+ str(q);
    return (p)
```

```python
def randdate():
    p= np.random.randint(low=1, high=31)
    q=np.random.randint(low=1, high=12)
    r=np.random.randint(low=2012, high=2017)
    p=str(p)+"/"+ str(q)+"/"+ str(r);
    return (p)

```
